```python
from datetime import datetime,timedelta
import time
import pandas as pd
import numpy as np
import talib as ta
import matplotlib.pyplot as plt
import seaborn as sb
import plotly.express as px
import plotly.graph_objects as go
import MetaTrader5 as mt5
import yfinance as yf
import mplfinance as mpf
from plotly.subplots import make_subplots
```

```python
import sklearn
from sklearn.tree import ExtraTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.neural_network import MLPRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.decomposition import PCA
#import tensorflow
```

```python
## Initialize connection to metatrader platform
mt5.initialize()
```

```python
#settings
#settings
symbol='USDCAD'
timeframe=mt5.TIMEFRAME_H1
start_pos=0
num_bars=5000
```

```python
#Extracting historical data
bars= mt5.copy_rates_from_pos(symbol,timeframe,start_pos,num_bars)
df=pd.DataFrame(bars)[['time','open','low','close','high']]
df['time']=pd.to_datetime(df['time'],unit='s')
df
```

```python
#to create direction column
def direction(num):
    if num <0:
        return 'sell'
    elif num >0:
        return 'buy'
direct=np.vectorize(direction)
```

```python
# to encode the direction
def label(char):
    if char =='sell':
        return 0
    elif char == 'buy':
        return 1
labels=np.vectorize(label)
```

## Feature engineering

```python
# Feature engineering
df['close_prev1']=df['close'].shift(1)
df['close_prev2']=df['close'].shift(2)
df['close_prev3']=df['close'].shift(3)
df['close_prev4']=df['close'].shift(4)
df['close_prev5']=df['close'].shift(5)
df['RSI']=ta.RSI(df['close'],10)
df['RSI_prev1']=ta.RSI(df['close_prev1'],10)
df['RSI_prev2']=ta.RSI(df['close_prev2'],10)
df['RSI_prev3']=ta.RSI(df['close_prev3'],10)
df['RSI_prev4']=ta.RSI(df['close_prev4'],10)
df['RSI_prev5']=ta.RSI(df['close_prev5'],10)

#df['ADX']=ta.ADX(df['high'],df['low'],df['close'],10)

df['SMA_10']=ta.SMA(df['close'],10)
df['SMA_10_prev1']=ta.SMA(df['close_prev1'],10)
df['SMA_10_prev2']=ta.SMA(df['close_prev2'],10)
df['SMA_10_prev3']=ta.SMA(df['close_prev3'],10)
df['SMA_10_prev4']=ta.SMA(df['close_prev4'],10)
df['SMA_10_prev5']=ta.SMA(df['close_prev5'],10)
df['SMA_20']=ta.SMA(df['close'],20)
df['SMA_20_prev1']=ta.SMA(df['close_prev1'],20)
df['SMA_20_prev2']=ta.SMA(df['close_prev2'],20)
df['SMA_20_prev3']=ta.SMA(df['close_prev3'],20)
df['SMA_20_prev4']=ta.SMA(df['close_prev4'],20)
df['SMA_20_prev5']=ta.SMA(df['close_prev5'],20)

df['direction']=direct(df['close'].diff(periods=1))
df['class']=labels(df['direction'])
df.dropna(inplace=True)
X=df.drop(columns=['close','time','open','high','low','direction','class'])
# for classification change Y to class instead of close values
Y=df['class']
X_train, X_test, y_train, y_test=train_test_split(X,Y,test_size=0.2,shuffle=False)
```

```python
X_train.shape,y_train.shape
```

## Dummy algorithm

```python
## normalize
std=StandardScaler()
X_train_std=std.fit_transform(X_train['close_prev1'].values.reshape(-1,1))
X_test_std=std.transform(X_test['close_prev1'].values.reshape(-1,1))
Tree_model=ExtraTreeRegressor()
Tree_model.fit(X_train_std,y_train)
y_pred=Tree_model.predict(X_test_std)
print('the r2-score is {}'.format(r2_score(y_test,y_pred)))
print('the mean squared error is {}'.format(mean_squared_error(y_test,y_pred)))

plt.figure(figsize=(12,8))
plt.plot(range(len(y_pred)),y_pred,'g-',label='predicted')
plt.plot(range(len(y_pred)),y_test,label='True') 
plt.legend()
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import plot_confusion_matrix
from sklearn.metrics import accuracy_score,f1_score
```

```python
X_train.drop(columns=['close_prev1','close_prev2','close_prev3','close_prev4','close_prev5'],inplace=True)
X_test.drop(columns=['close_prev1','close_prev2','close_prev3','close_prev4','close_prev5'],inplace=True)
```

```python
y_train=y_train.astype('int')
y_test=y_test.astype('int')
```

### KNN Classifier

```python
std=StandardScaler()
X_train_std=std.fit_transform(X_train)
X_test_std=std.transform(X_test)
knn=KNeighborsClassifier()
knn.fit(X_train,y_train)
y_pred=knn.predict(X_test)
print('the accuracy is {}'.format(accuracy_score(y_test,y_pred)))
print('the f1_score is {}'.format(f1_score(y_test,y_pred)))
plot_confusion_matrix(knn,X_test,y_test);
```
